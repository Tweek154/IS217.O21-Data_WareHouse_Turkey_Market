```python
import pandas as pd
import numpy as np
```

```python
Airlane_DF = pd.read_csv('Airlane_Passenger_SatisFaction.csv')
Airlane_DF = Airlane_DF.loc[:, ~Airlane_DF.columns.str.contains('^Unnamed')]
Airlane_DF
```

```python
numeric_cols = Airlane_DF.select_dtypes(include ='number').columns
numeric_cols
```

```python
stats = pd.DataFrame(Airlane_DF[numeric_cols].describe().T)
stats
```

```python
for col in Airlane_DF[numeric_cols].columns:
    stats.loc[col, 'mode'] = Airlane_DF[col].mode().iloc[0]  # Using iloc[0] to get the first mode if multiple
    stats.loc[col, 'range'] = Airlane_DF[col].max() - Airlane_DF[col].min()
    stats.loc[col, 'variance'] = Airlane_DF[col].var()  # Correcting the spelling of 'variance'
display(stats)
```

```python
#Cho biết loại hình dịch vụ nào được ưa chuộng nhất?
import matplotlib.pyplot as plt
import seaborn as sns
type_counts = Airlane_DF['Class'].value_counts()
plt.figure(figsize=(15,10))
plt.title("The number of customers depend on class travel")
sns.barplot(x = type_counts[:10].values, y = type_counts[:10].index, orient = "h")
plt.xlabel("Number of flight")
plt.ylabel("Class")
plt.show()
```

```python
X = Airlane_DF.drop('id', axis = 1)
X.info()
```

```python
print("Number of rows: ", len(X.index))
X.drop_duplicates(inplace=True)
```

```python
print("Number of rows after drop of duplicates: ", len(X.index))
```

```python
X.info()
```

```python
X = X.dropna(subset=['Arrival Delay in Minutes'])
X
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python
original_values = {}

for column in X.columns:
    if X[column].dtype == 'object':
        original_values[column] = X[column].copy() 

        # Fit và transform dữ liệu của cột
        X[column] = encoder.fit_transform(X[column])
```

```python
from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()
```

```python
features_to_scale = ['Age','Flight Distance', 'Departure Delay in Minutes', 'Arrival Delay in Minutes']
X[features_to_scale] = std_scaler.fit_transform(X[features_to_scale])
X.head(5)
```

```python
X
```

### Ứng dụng mô hình thuật toán khai thác dữ liệu

```python
accuracy = {}
f1 = {}
recall = {}
precision = {}
tpr = {}
fpr = {}
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
```

```python
X, X_test = train_test_split(X,test_size = 0.2, random_state=42)
print(X.shape, X_test.shape)
```

```python
sample = X
y_sample = sample['satisfaction']
X_sample = sample.drop('satisfaction', axis = 1)
X_train, X_validate, y_train, y_validate = train_test_split(X_sample, y_sample, random_state=42)
print(X_train.shape, y_train.shape, X_validate.shape, y_validate.shape)
```

## Thuật toán cây quyết định

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
```

```python
dtc = DecisionTreeClassifier(random_state=42)
parameters = [{'criterion':["gini","entropy"], "max_depth" : [5,10,15,30]}]
grid = GridSearchCV(dtc,parameters, verbose = 5, n_jobs = -1)
grid.fit(X_train,y_train)
```

```python
print("Best parameters scores:", grid.best_params_)
print("Train score:", grid.score(X_train,y_train))
print("Validation score:", grid.score(X_validate,y_validate))
```

```python
dtc.fit(X_train,y_train)
print("Train score:", dtc.score(X_train,y_train))
print("Validation score:", dtc.score(X_validate,y_validate))
```

```python
pd.DataFrame(grid.cv_results_).sort_values(by="rank_test_score")
```

```python
y_pred = dtc.predict(X_validate)
confmat = confusion_matrix(y_true = y_validate, y_pred = y_pred)
index = ['Actual Neutral or Dissatisfied', 'Actual Satisfied']
columns = ['Predicted Neutral or Dissatisfied', 'Predicted Satisfied']
conf_matrix = pd.DataFrame(data = confmat, columns = columns, index = index)
print(confmat)
plt.figure(figsize=(10,8))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="YlGnBu")
plt.title("Confusion matrix - Decision tree")
plt.show()
```

```python
importances = pd.DataFrame(np.zeros((X_train.shape[1],1)), columns = ["importance"], index = X_train.columns)
importances.iloc[:,0] = dtc.feature_importances_
importances = importances.sort_values(by="importance", ascending = False)[:10]
plt.figure(figsize=(15,10))
sns.barplot(x="importance", y = importances.index, data = importances)
plt.show()
```

```python
from sklearn.tree import plot_tree

fig, ax = plt.subplots(figsize=(20,10))
plot_tree(dtc,max_depth=4, fontsize=10, feature_names = X_train.columns.to_list(),class_names = True, filled = True)
plt.show()
```

```python
from sklearn.metrics import f1_score, precision_score, recall_score, accuracy_score
```

```python
f1['Decision Tree'] =  f1_score(y_validate,y_pred,average='macro')
precision['Decision Tree'] = precision_score(y_validate,y_pred,average='macro')
recall['Decision Tree'] = recall_score(y_validate,y_pred,average='macro')
accuracy['Decision Tree'] = accuracy_score(y_validate,y_pred)
print(f1['Decision Tree'])
print(precision['Decision Tree'])
print(recall['Decision Tree'])
print(accuracy['Decision Tree'])
```

```python
plt.plot(dc_fpr, dc_tpr, marker = '.', label = 'Decision Tree (AUROC = %0.3f)' % dc_auc)
plt.plot(rf_fpr, rf_tpr, marker = '.', label = 'Random Forest (AUROC = %0.3f)' % rf_auc)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend()
plt.show()
```
